```python

from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
```

```python
url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'5000',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': '8fbbabfe-7e32-4ef2-8cd8-b0a56ae98423',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
symbolNameList = []
for entry in data['data']:
  symbolNameList.append((entry['symbol'], entry['name']))
```

```python
with open('coinNameMapping.txt', 'w') as f:
  for item in symbolNameList:
    f.write(str(item)+"\n")
```

```python
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive2 = GoogleDrive(gauth)
```

```python
def findCoin (coin):
  found = ""
  extractCoin = coin[:-4].upper() if coin != "usdtinr" else coin[:-3].upper()
  for item in symbolNameList:
    if (item[0] == extractCoin):
      found = (coin, item[1])
  return found

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

drivepath = "/content/drive/Shareddrives/ProjectSharedDrive/wazirxCSVData"
driveDAILYpath = "/content/drive/Shareddrives/ProjectSharedDrive/wazirxDAILY"


tickers = open(drivepath+"/tickers" ,"r")
tfile = json.load(tickers)
tickers.close()

coins = list(tfile.keys())
USDTcoins = [s for s in coins if "usdt" in s]
```

```python
usdtCoinName = []
for coin in USDTcoins:
  name = findCoin(coin)
  if name != '':
    usdtCoinName.append(name)

usdtCoinName.append(("bchsvusdt", "Bitcoin Cash SV"))
usdtCoinName.append(("iotausdt", "MIOTA"))
usdtCoinName.append(("gxsusdt", "GXChain"))
```

```python
with open('USDTCoinName.txt', 'w') as f:
  for item in usdtCoinName:
    f.write(str(item)+"\n")
```
