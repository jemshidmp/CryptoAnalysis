```python
cols = ['data']
# Crypto_lists.json file is created from "https://coinmarketcap.com/api/" Website
# API Link:  https://coinmarketcap.com/api/documentation/v1/#operation/getV1CryptocurrencyListingsLatest
file_name = '/content/drive/Shareddrives/ProjectSharedDrive/Crypto_lists.json'

# Opening JSON file
f = open(file_name)

# returns JSON object as
# a dictionary
data = json.load(f)

# Counter to take top 25 Cryptocurrency
counter = 0

with open('/content/drive/Shareddrives/ProjectSharedDrive/top25_cryptocoins_list.jsonl', 'w') as outfile:
    for entry in data['data']:
        counter = counter + 1
        my_dict={}
        my_dict['name']=entry.get('name')
        my_dict['symbol']=entry.get('symbol').lower()+ 'usdt'
        my_dict['cmc_rank']=entry.get('cmc_rank')
        json.dump(my_dict, outfile)

        if(counter == 25):
          break
        outfile.write('\n')
```

```python
cryptofile = "/content/drive/Shareddrives/ProjectSharedDrive/top25_cryptocoins_list.jsonl"


_crypt_raw = spark.read.json(cryptofile)
_crypt = _crypt_raw.withColumn("name", col("name")) \
          .withColumn("symbol", col("symbol")) \
          .withColumn("cmc_rank", col("cmc_rank")).orderBy("cmc_rank")
#_crypt.show()
top10=[]
j=0
for i in _crypt.collect():
   top10.append(i[1])
   j+=1
   if(j==10):
     break

# top10_df = _crypt.select("name")
# top10 = top25_df.collect()
top10
```
